### TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


<details><summary>Click here for a hint</summary>

<p>

````python
# 1) Aggregate sales by year (sum over months)
sales_by_year = df.groupby('Year', as_index=True)['Automobile_Sales'].sum()

# 2) Build the plot using pandas
ax = sales_by_year.plot(kind='line', marker='o', figsize=(12, 5))

# 3) Axis labels & title
ax.set_xlabel('Year')
ax.set_ylabel('Total Automobile Sales')
ax.set_title('Automobile Sales during Recession')

# 4) Show every year on the x-axis
ax.set_xticks(sales_by_year.index)
ax.set_xticklabels(sales_by_year.index, rotation=45)

# 5) Highlight/annotate recession years
recession_years = df.loc[df['Recession'] == 1, 'Year'].drop_duplicates().tolist()

# Light vertical guides for all recession years
for y in recession_years:
    ax.axvline(y, color='red', linestyle='--', alpha=0.15)

# Annotations for at least two recession years
for y in recession_years[:2]:
    if y in sales_by_year.index:
        ax.annotate('Recession',
                    xy=(y, sales_by_year.loc[y]),
                    xytext=(y, sales_by_year.loc[y] * 1.05),
                    ha='center',
                    arrowprops=dict(arrowstyle='->', color='red'))

plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# ---- Build a time axis (works if you have Year & Month in the file) ----
# If Month is a name like "January", map it to a number; else this line is a no-op.
month_map = {m:i for i,m in enumerate(
    ["January","February","March","April","May","June",
     "July","August","September","October","November","December"], 1)}
if 'Month' in df.columns and df['Month'].dtype == object:
    df['Month_num'] = df['Month'].map(month_map)
else:
    df['Month_num'] = df.get('Month', 1)

df['Date'] = pd.to_datetime(dict(year=df['Year'], month=df['Month_num'], day=1))

# ---- Filter to recession period(s) ----
rec = df[df['Recession'] == 1]

# ---- Aggregate and pivot: one column per Vehicle Type ----
by_type = (rec.groupby(['Year','Vehicle_Type'])['Automobile_Sales']
             .sum().unstack('Vehicle_Type').sort_index())
by_type.plot(figsize=(12,6), lw=2)
plt.xticks(by_type.index, rotation=45)  # show every year


# ---- Plot lines for each vehicle type ----
plt.figure(figsize=(12,6))
ax = by_type.plot(lw=2)
ax.set_title('Sales Trends by Vehicle Type During Recession Periods')
ax.set_xlabel('Date')
ax.set_ylabel('Automobile Sales')
ax.grid(alpha=.3)
ax.legend(title='Vehicle Type', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

````python
import matplotlib.pyplot as plt
import seaborn as sns

# If your column names differ, tweak these:
sales_col   = "Automobile_Sales"
type_col    = "Vehicle_Type"
recess_col  = "Recession"          # 1 = recession, 0 = not

# Nice label for plotting
df["Period"] = df[recess_col].map({1: "Recession", 0: "Non-recession"})

# Aggregate sales by Vehicle Type and Period
vt = (df.groupby([type_col, "Period"], as_index=False)[sales_col]
        .sum()
        .sort_values([type_col, "Period"]))

plt.figure(figsize=(11,6))
sns.barplot(data=vt, x=type_col, y=sales_col, hue="Period")
plt.title("Automobile Sales by Vehicle Type: Recession vs Non-recession")
plt.xlabel("Vehicle type")
plt.ylabel("Total sales")
plt.xticks(rotation=20, ha="right")
plt.legend(title="")
plt.tight_layout()
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import matplotlib.pyplot as plt
import pandas as pd

# --- pick the GDP column name safely ---
gdp_col = 'GDP' if 'GDP' in df.columns else ('GDP(in billions)' if 'GDP(in billions)' in df.columns else None)
assert gdp_col is not None, "Couldn't find a GDP column in the dataframe."

# --- ensure a proper time axis ---
if 'Date' not in df.columns:
    # if Month exists, build a month-level Date; else use Year 1st of Jan
    if 'Month' in df.columns:
        month_map = {m:i for i,m in enumerate(
            ["January","February","March","April","May","June",
             "July","August","September","October","November","December"], 1)}
        month_num = df['Month'].map(month_map) if df['Month'].dtype == object else df['Month']
        df['Date'] = pd.to_datetime(dict(year=df['Year'], month=month_num, day=1))
    else:
        df['Date'] = pd.to_datetime(df['Year'].astype(int).astype(str) + "-01-01")

# --- split & aggregate (mean to avoid repeated rows per month/year) ---
rec_gdp  = (df[df['Recession'] == 1]
            .groupby('Date')[gdp_col].mean().sort_index())

nonrec_gdp = (df[df['Recession'] == 0]
              .groupby('Date')[gdp_col].mean().sort_index())

# --- plot: two subplots side-by-side ---
fig = plt.figure(figsize=(14,5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2, sharey=ax1)

rec_gdp.plot(ax=ax1, color='tab:red', lw=2)
ax1.set_title('GDP During Recession')
ax1.set_xlabel('Date'); ax1.set_ylabel('GDP')
ax1.grid(alpha=.3)

nonrec_gdp.plot(ax=ax2, color='tab:blue', lw=2)
ax2.set_title('GDP During Non-Recession')
ax2.set_xlabel('Date')
ax2.grid(alpha=.3)

plt.tight_layout()
plt.show()

# Save if needed
plt.savefig("Subplot.png", dpi=150, bbox_inches='tight')
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- keep only non-recession months ---
nonrec = df[df['Recession'] == 0].copy()

# Make sure month labels are ordered correctly
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

if nonrec['Month'].dtype != 'O':  # numeric months -> names
    nonrec['Month'] = nonrec['Month'].astype(int).map(dict(zip(range(1,13), month_order)))

nonrec['Month'] = pd.Categorical(nonrec['Month'],
                                 categories=month_order, ordered=True)

# Aggregate by month: average sales & average seasonality weight
grp = (nonrec.groupby('Month', observed=False)
              .agg(avg_sales=('Automobile_Sales','mean'),
                   avg_weight=('Seasonality_Weight','mean'))
              .reset_index())

# Bubble sizes proportional to Seasonality_Weight (area scaling)
sizes = (grp['avg_weight'] / grp['avg_weight'].max()) * 2000  # adjust 2000 if you want bigger/smaller bubbles

# --- bubble plot ---
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(grp['Month'], grp['avg_sales'], s=sizes, alpha=0.6,
           color='tab:blue', edgecolor='k')

ax.set_title('Seasonality impact on Automobile Sales')
ax.set_xlabel('Month')
ax.set_ylabel('Average Automobile Sales (Non-recession years)')
ax.grid(alpha=0.3)
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("Bubble.png", dpi=150, bbox_inches='tight')
plt.show()
````

 #### From the data, develop a scatter plot to identify if there a correlation between consumer confidence and automobile sales during recession period?

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ---- keep only recession rows ----
rec = df[df['Recession'] == 1].copy()

# ---- 1) Consumer confidence vs. automobile sales (during recessions) ----
x_cc = 'Consumer_Confidence'  # column name in this dataset
y_sales = 'Automobile_Sales'

r_cc = rec[[x_cc, y_sales]].corr().iloc[0, 1]  # Pearson r

plt.figure(figsize=(8,5))
plt.scatter(rec[x_cc], rec[y_sales], alpha=0.7, edgecolor='k')
plt.title('Consumer Confidence and Automobile Sales during Recessions')
plt.xlabel('Consumer Confidence')
plt.ylabel('Automobile Sales')
plt.grid(alpha=0.25)
plt.text(0.02, 0.95, f'Pearson r = {r_cc:.2f}', transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round', fc='white', ec='gray', alpha=0.8))
plt.tight_layout()
plt.show()

# ---- 2) Average vehicle price vs. automobile sales (during recessions) ----
# Be tolerant to possible column-name variants
price_candidates = ['Price', 'Average_Vehicle_Price', 'Avg_Vehicle_Price', 'AveragePrice']
price_col = next((c for c in price_candidates if c in df.columns), None)
assert price_col is not None, "Couldn't find a price column. Check the dataframe columns."

r_price = rec[[price_col, y_sales]].corr().iloc[0, 1]

plt.figure(figsize=(8,5))
plt.scatter(rec[price_col], rec[y_sales], alpha=0.7, color='tab:orange', edgecolor='k')
plt.title('Relationship between Average Vehicle Price and Sales during Recessions')
plt.xlabel('Average Vehicle Price')
plt.ylabel('Automobile Sales')
plt.grid(alpha=0.25)
plt.text(0.02, 0.95, f'Pearson r = {r_price:.2f}', transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round', fc='white', ec='gray', alpha=0.8))
plt.tight_layout()
plt.show()
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
 <br>Analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period
 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

<p>
    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>
</details>


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---------- column lookups (robust to small name variations) ----------
veh_candidates = ['Vehicle_Type','vehicle_type','Vehicle type','Type']
sales_candidates = ['Automobile_Sales','automobile_sales','Sales','sales']
ur_candidates    = ['Unemployment_Rate','unemployment_rate','Unemployment rate','unemployment']
date_candidates  = ['Year','year','Date','date','Month']

def pick(cols, candidates):
    return next((c for c in candidates if c in cols), None)

veh_col  = pick(df.columns, veh_candidates)
sales_col= pick(df.columns, sales_candidates)
ur_col   = pick(df.columns, ur_candidates)
x_col    = pick(df.columns, date_candidates)

assert veh_col and sales_col and ur_col and x_col, f"Missing needed columns. Have: {list(df.columns)}"

# make a year column if x is 'Date'
if x_col.lower() == 'date':
    df['_Year_'] = pd.to_datetime(df[x_col]).dt.year
    x_col = '_Year_'

# ---------- filter to recession period ----------
recess = df[df['Recession'] == 1].copy()

# focus on the 3 types mentioned (fallback to top-3 if names differ)
wanted = ['superminicar','smallfamilycar','mediummnicar','mediumminicar','medium family car']
# match ignoring case
present_map = {v.lower(): v for v in recess[veh_col].dropna().unique()}
use_types = [present_map[w] for w in wanted if w in present_map]
if len(use_types) == 0:
    # fallback: take three most frequent types
    use_types = recess[veh_col].value_counts().head(3).index.tolist()

data_plot = recess[recess[veh_col].isin(use_types)].copy()

# ---------- plot ----------
sns.set(style='whitegrid')
fig, ax1 = plt.subplots(figsize=(10,5))

# sales by vehicle type
sns.lineplot(
    data=data_plot.sort_values([x_col]),
    x=x_col, y=sales_col, hue=veh_col, marker='o',
    hue_order=use_types, ax=ax1
)
ax1.set_xlabel('Year')
ax1.set_ylabel('Automobile Sales (units)')
ax1.set_title('Effect of Unemployment Rate on Vehicle Type and Sales (Recession)')

# unemployment rate (average across types per year) on a secondary axis
ur_series = recess.groupby(x_col, as_index=False)[ur_col].mean()
ax2 = ax1.twinx()
sns.lineplot(data=ur_series, x=x_col, y=ur_col, color='black',
             linestyle='--', marker='s', label='Unemployment rate', ax=ax2)
ax2.set_ylabel('Unemployment rate (%)')
ax2.legend(loc='upper left', bbox_to_anchor=(1.01, 1), frameon=True)

plt.tight_layout()
plt.show()

# ---------- correlation (unemployment vs sales) by vehicle type ----------
corr_by_type = (
    data_plot.groupby(veh_col)
             .apply(lambda g: g[[ur_col, sales_col]].corr(method='pearson').iloc[0,1])
             .rename('corr(unemp, sales)')
             .sort_values()
)
print('Correlation between unemployment rate and sales (recession):')
display(corr_by_type.to_frame())
````
